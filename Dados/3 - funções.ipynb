```python
import numpy as np
```

```python
array = np.array([1,2,3,4,5])
array
```

```python
# Adição

soma = np.add(array,10)
soma
```

```python
# Subtração

sub = np.subtract(array,1)
sub
```

```python
# Multiplicação

multi = np.multiply(array,2)
multi
```

```python
#Divisão 

div = np.divide(array, 2)
div
```

```python
#Potência

potencia = np.pow(array, 2)
potencia
```

Funções estatísticas

```python
#Média

media = np.mean(array)
media
```

```python
#Desvio Padrão

dev_padrao = np.std(array)
dev_padrao
```

```python
#Comparação de valores do Array

maior_que_2 = array > 2 
maior_que_2
```

```python
# Combinando Funções

resposta = np.sqrt(array) + np.mean(array)
resposta
```
