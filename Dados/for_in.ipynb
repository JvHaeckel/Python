```python
# Imprima os números de 1 a 10 usando for.

for i in range(1,6):
    print(i)
```

```python
# Some os valores da lista abaixo sem usar sum().

numeros = [2, 4, 6, 8]

soma = 0 

for i in numeros:
    soma+= i   # soma = soma + i
    
print(soma)
```

```python
# Conte quantos elementos existem na lista sem usar len().

valores = [10, 20, 30, 40, 50]

cont = 0

for i in valores: 
    cont = cont + 1
    
print(cont)
```

```python
# Imprima apenas os números maiores que 10. 
dados = [3, 15, 7, 20, 1, 11]


for i in dados:
    
    if i > 10: print(f"Maior que 10: {i}")
    else: print("Menor que 10: {i}")
```

```python
# Percorra os números de 1 a 10 e diga se são pares ou ímpares.


for i in range(1,11):
    
    if i % 2 == 0: print(f"Par  : {i}")
    else: print(f"Ímpar: {i}" )
```
